```python
from functions.io_data import read_data, write_data, convert_lab_data_to_pil_image
from scipy.signal import convolve2d
import numpy as np
```

## 1. Gibbs Sampling Denoising

```python
# ---------------------- Parameters ----------------------
SEED      = 100
LABELS    = [-1,1] 
SIGMA     = 1   # Noise for Gaussian in Local potential
J         = 1   # Coupling strength in Pairwise potential
#BURN_IN   = 5 # Burn-in period
MAX_STEPS = 10  # Maximum number of steps
np.random.seed(SEED)

# ----------------------- Load Data ----------------------
def load(filename):
    # We put (is_RGB = True) to obtain the image in LAB space
    data, img, pil_img = read_data(filename, is_RGB=True, save=False)
    img_data = np.squeeze(img, axis=2)

    print("Data shape: %s" % str(data.shape))
    print("Img shape : %s" % str(img.shape))
    return data, img_data, pil_img

# -------------------- Initialization --------------------
def initialize(X):
    # Initialize variables to the original image
    new_X = np.zeros_like(X)
    new_X[X == 0] = -1
    new_X[X == 255] = 1
    return new_X

def sample(probs):
    # Sample an index when given a list of probabilities (can be unnormalized)
    norm_factor = sum(probs)
    norm_probs = probs / norm_factor
    idx = np.argmax(np.random.multinomial(1, norm_probs))
    return idx

# -------------------- Gibbs Sampling --------------------
def denoise_gibbs_sampling(filename):
    data, img_data, pil_img = load(filename)
    X = initialize(img_data)
    X_neg = np.ones_like(X) * -1
    X_pos = np.ones_like(X)
    H, W = X.shape

    kernel_top   = [[0,1,0],[0,0,0],[0,0,0]] # Retrieve the top
    kernel_left  = [[0,0,0],[1,0,0],[0,0,0]] # Retrieve the left
    kernel_bot   = [[0,0,0],[0,0,0],[0,1,0]] # Retrieve the bot
    kernel_right = [[0,0,0],[0,0,1],[0,0,0]] # Retrieve the right

    for step in range(MAX_STEPS):
        if step % 5 == 0: print("Step %d" % step)
        X_top   = convolve2d(X, kernel_top, mode='same')
        X_left  = convolve2d(X, kernel_left, mode='same')
        X_bot   = convolve2d(X, kernel_bot, mode='same')
        X_right = convolve2d(X, kernel_right, mode='same')
    
        # Pairwise potentials
        neg_pairw_potentials = np.exp(J * X_neg * X_top) * np.exp(J * X_neg * X_left) * \
                               np.exp(J * X_neg * X_bot) * np.exp(J * X_neg * X_right)
        pos_pairw_potentials = np.exp(J * X_pos * X_top) * np.exp(J * X_pos * X_left) * \
                               np.exp(J * X_pos * X_bot) * np.exp(J * X_pos * X_right)
        
        # Local potentials
        neg_local_potentials = (1. / np.sqrt(2 * np.pi * SIGMA**2)) * \
                               np.exp(-(X - (-1))**2 / (2 * SIGMA**2))
        pos_local_potentials = (1. / np.sqrt(2 * np.pi * SIGMA**2)) * \
                               np.exp(-(X - (1))**2 / (2 * SIGMA**2))
        
        # Combine potentials and normalize
        neg_potentials = neg_pairw_potentials * neg_local_potentials
        pos_potentials = pos_pairw_potentials * pos_local_potentials
        norm_factor = neg_potentials + pos_potentials
        neg_prob = neg_potentials / norm_factor
        pos_prob = pos_potentials / norm_factor
        probs = np.stack([neg_prob, pos_prob], axis=2)
        
        # Sample from the probabilities
        for i in range(H):
            for j in range(W):
                sampled_label = LABELS[sample(probs[i,j,:])]
                X[i,j] = sampled_label
    
    # Form binary image after some iterations
    img_X = np.zeros_like(X)
    img_X[X == -1] = 0
    img_X[X == 1] = 255
    
    return img_X

img_X = denoise_gibbs_sampling('a1/2_noise.txt')

# Visualize
from PIL import Image
H, W = img_X.shape
pil_img = Image.fromarray(np.reshape(img_X, [H, W]).astype(np.uint8), 'L')
pil_img
```

## 2. Variational Inference Denoising

```python
# ---------------------- Parameters ----------------------
SEED      = 100
W         = 1  # Coupling strength in Pairwise
C         = 1  # Sigma for Gaussian in Local
MAX_STEPS = 10 # Maximum number of steps
np.random.seed(SEED)

# ----------------------- Load Data ----------------------
def load(filename):
    # We put (is_RGB = True) to obtain the image in LAB space
    data, img, pil_img = read_data(filename, is_RGB=True, save=False)
    img_data = np.squeeze(img, axis=2)

    print("Data shape: %s" % str(data.shape))
    print("Img shape : %s" % str(img.shape))
    return data, img_data, pil_img

# ---------------- Updates to Parameters -----------------
def update_Q(Q, img, mus):
    kernel_neighbors = [[0,W,0],[W,0,W],[0,W,0]]
    m = convolve2d(mus, kernel_neighbors, mode='same')
    a = m + 0.5 * (L(img, 1) - L(img, -1))
    Q[:,:,0] = sigmoid(-2 * a)
    Q[:,:,1] = sigmoid(2 * a)

def update_mus(mus, Q):
    mus[:,:] = -1 * Q[:,:,0] + 1 * Q[:,:,1]
    
def sigmoid(x):
    return 1. / (1. + np.exp(-x))

def L(Y, X):
    return (1. / np.sqrt(2 * np.pi * C**2)) * np.exp(-(Y - X)**2 / (2 * C**2))

# -------------------- Initialization --------------------
def initialize(X):
    # Initialize variables to the original image
    new_X = np.zeros_like(X)
    new_X[X == 0] = -1
    new_X[X == 255] = 1
    
    # Initialize mus
    mus = np.zeros_like(X)
    
    # Initialize Q
    H, W = X.shape
    Q = np.zeros((H, W, 2))
    
    update_Q(Q, new_X, mus)
    
    return new_X, mus, Q

# ---------------- Variational Inference -----------------
def denoise_variational_inference(filename):
    data, img_data, pil_img = load(filename) 
    X_img, mus, Q = initialize(img_data)

    for step in range(MAX_STEPS):
        if step % 5 == 0: print("Step %d" % step)
        update_Q(Q, X_img, mus)
        update_mus(mus, Q)
    
    # Form binary image
    img_final = np.zeros_like(X_img)
    img_final[np.argmax(Q, axis=2) == 0] = 0
    img_final[np.argmax(Q, axis=2) == 1] = 255
    
    return img_final

img_final = denoise_variational_inference('a1/3_noise.txt')
pil_img = Image.fromarray(img_final.astype(np.uint8), 'L')
pil_img
```

## 3. Expectation-Maximization Segmentation

```python
# ---------------------- Parameters ----------------------
SEED      = 100
K         = 2  # Number of components
D         = 3  # Number of dimensions in LAB space
MAX_STEPS = 50 # Maximum number of EM steps
np.random.seed(SEED)

# ----------------------- Load Data ----------------------
def load(filename):
    # We put (is_RGB = True) to obtain the image in LAB space
    data, img, pil_img = read_data(filename, is_RGB=True, save=False)
    lab_data = data[:,2:]

    print("Data shape: %s" % str(data.shape))
    print("Img shape : %s" % str(img.shape))

    return data, lab_data

# -------------------- Initialization --------------------
def initialize(X):
    N = X.shape[0]
    mus, sigmas, pis = [], [], []
    
    # Initialize mus to K randomly selected data values
    for idx in np.random.choice(N, K):
        mus.append(X[idx,:])

    # Initialize sigmas to covs of K randomly selected subsets of the data
    for k in range(K):
        idxs = np.random.choice(N, 100)
        sigma = np.cov(X[idxs,:], rowvar=False)
        sigmas.append(sigma)

    # Initialize pis with uniform distribution from 0 to 1
    unnorm_probs = [np.random.rand() for k in range(K)]
    for unnorm_prob in unnorm_probs:
        pis.append(unnorm_prob / sum(unnorm_probs))
    
    return np.array(mus), np.array(sigmas), np.array(pis)

# ------------------- Expectation Step -------------------
def multivariate_normal_pdf(X, mu, sigma):
    """
    X_dist = []
    for i in range(X.shape[0]):
        const = 1. / np.sqrt(np.linalg.det(2 * np.pi * sigma))
        expon = np.exp(-0.5 * np.dot(np.dot(X[i,:] - mu, np.linalg.inv(sigma)), (X[i,:] - mu).T))
        X_dist.append(const * expon)
    return np.array(X_dist)
    """
    from scipy.stats import multivariate_normal
    return multivariate_normal.pdf(X, mean=mu, cov=sigma)

def responsibility(X, mus, sigmas, pis):
    # Calculates the responsibility of the k-th component
    pdf_times_pis = np.array([pis[i] * multivariate_normal_pdf(X, mus[i], sigmas[i]) \
                              for i in range(K)])     # (2,N)
    pdf_times_pis_sum = np.sum(pdf_times_pis, axis=0) # (N)
    responsibility = np.array([pdf_times_pis[i]/pdf_times_pis_sum \
                               for i in range(K)])    # (2,N)
    return responsibility.T # (N,2)

def E_step(X, mus, sigmas, pis):
    responsibilities = responsibility(X, mus, sigmas, pis)
    return responsibilities

# ------------------- Maximization Step ------------------
def M_step(X, resp, mus, sigmas, pis):
    new_mus, new_sigmas, new_pis = [], [], []
    N = X.shape[0]
    
    for k in range(K):
        N_k = np.sum(resp[:,k])
        new_mu = np.dot(resp[:,k], X) / N_k
        new_sigma = np.array([resp[i,k] * \
                              np.dot((X[i:i+1,:] - new_mu).T, 
                                     (X[i:i+1,:] - new_mu)) \
                              for i in range(N)])   # (N,3,3)
        new_sigma = np.sum(new_sigma, axis=0) / N_k # (3,3)
        new_pi = N_k / N
        
        new_mus.append(new_mu)
        new_sigmas.append(new_sigma)
        new_pis.append(new_pi)
        
    return np.array(new_mus), np.array(new_sigmas), np.array(new_pis)

# ---------------- Evaluate Log Likelihood ---------------
def log_likelihood(X, mus, sigmas, pis):
    pdf_times_pis = [pis[k] * multivariate_normal_pdf(X, mus[k], sigmas[k]) \
                     for k in range(K)]       # (2,N)
    l = np.log(np.sum(pdf_times_pis, axis=0)) # (N)
    return np.sum(l)

# -------------------------- EM --------------------------
def EM(X):
    mus, sigmas, pis = initialize(X)
    for i in range(MAX_STEPS):
        resp = E_step(X, mus, sigmas, pis)
        mus, sigmas, pis = M_step(X, resp, mus, sigmas, pis)
        l = log_likelihood(X, mus, sigmas, pis)

        # Convergence checking
        should_break = False
        if i != 0 and l - old_l < 1e-5: should_break = True
        if i != 0:
            mus_converged = np.sum(np.abs(old_mus - mus)) < 1e-3
            sigmas_converged = np.sum(np.abs(old_sigmas - sigmas)) < 1e-3
            pis_converged = np.sum(np.abs(old_pis - pis)) < 1e-3
            if mus_converged and sigmas_converged and pis_converged: should_break = True
        if should_break: 
            print('Ended on Step %d, Log-likelihood: %f' % (i,l))
            return l, mus, sigmas, pis
        
        # Parameters from previous step (used for convergence checking)
        old_mus = mus
        old_sigmas = sigmas
        old_pis = pis
        old_l = l
    
        if i % 5 == 0: print('Step %d, Log-likelihood: %f' % (i,l))
        if i == MAX_STEPS - 1: return l, mus, sigmas, pis

def EM_with_reruns(X, reruns=1):
    # Run multiple EMs and choose the one with best log-likelihood
    assert(reruns > 0)
    print(" *** RUN %d ***" % 1)
    best_l, best_mus, best_sigmas, best_pis = EM(X)
    for i in range(reruns - 1):
        print(" *** RUN %d ***" % i+2)
        l, mus, sigmas, pis = EM(X)
        if l > best_l:
            best_mus, best_sigmas, best_pis = mus, sigmas, pis
    return best_mus, best_sigmas, best_pis

animal = 'zebra'
data, lab_data = load('a2/%s.txt' % animal)
mus, sigmas, pis = EM_with_reruns(lab_data, reruns=1)
resp = responsibility(lab_data, mus, sigmas, pis)
labels = np.argmax(resp, axis=1)
foreground_idx = 0 if np.sum(labels == 0) < np.sum(labels == 1) else 1

import matplotlib.pyplot as plt

data2 = np.copy(data)
data2[:,2:] *= np.expand_dims((labels == foreground_idx).astype(np.float32), axis=1)
write_data(data2, "a2/%s_foreground.txt" % animal)
_, _, pil_img = read_data("a2/%s_foreground.txt" % animal, is_RGB=True, save=False)

data2 = np.copy(data)
data2[:,2:] *= np.expand_dims((labels == 1 - foreground_idx).astype(np.float32), axis=1)
write_data(data2, "a2/%s_background.txt" % animal)
_, _, pil_img2 = read_data("a2/%s_background.txt" % animal, is_RGB=True, save=False)

print(np.sum(labels == 0))
print(np.sum(labels == 1))

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(pil_img)
plt.subplot(1,2,2)
plt.imshow(pil_img2)
plt.show()
```
